### 1. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import numpy as np
```

### 2. Load Data

```python
df = pd.read_csv('../data/processed/fake_news_combined.csv')
df.head(5)
```

### 3. Basic Info

```python
sns.countplot(x='label', data=df)
plt.title("Class Distribution (Fake vs Real)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Text Length Analysis
df['text_length'] = df['text'].apply(lambda x: len(str(x).split()))
sns.histplot(data=df, x='text_length', hue='label', bins=50, kde=True)
plt.title("Distribution of Article Length by Label")
plt.show()
```

```python
# 7. Check Duplicates
print("Total duplicate rows:", df.duplicated().sum())
```

```python
# 8. Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# save cleaned data
df.to_csv("../data/processed/fake_news_combined_cleaned.csv", index=False)
```
